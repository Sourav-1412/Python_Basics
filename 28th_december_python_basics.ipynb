# **TOPIC: String**

**ANSWER 1**

```python
# take user input as string
string = input()

# printing length of string variabLE
print(f"length of string is: {len(string)}")
```

**ANSWER 2**

```python
# taking user input
string = input("Type the sentence: ")
# creating an string variable containing all vowels in alphabets
vowels = "AaEeIiOoUu"

# using loop to check the vowels in user given sentence
time = 0
for s in string:
    if s in vowels:
        time = time+1
    
print(f"The number of time vowels come: {time}")
```

**ANSWER 3**

```python
# Given a string
string = "hello sourav"

# reversing the order of characters
new_string = string[len(string)::-1]

# printing the reversed string
print(new_string)
```

**ANSWER 4**

```python
# taking user input
string = input("Type the sentence: ")

# reversing the order of characters
reverse = string[len(string)::-1]

# checks if it is a palindrome or not
if(string == reverse):
    print("Yes, this string is palindrome")
else:
    print("No, this string is not palindrome")
```
